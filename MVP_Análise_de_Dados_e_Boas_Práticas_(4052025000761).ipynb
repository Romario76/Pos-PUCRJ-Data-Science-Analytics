# Descrição do Problema

O conjunto com dados contém resultados reais de testes em fluidos aplicados em máquinas de grande porte utilizadas nos processos de movimentação e extração de materiais em áreas de mineração. O objetivo principal é classificar o estado geral da amostra utilizada, com base nestes dados.


## Hipóteses do Problema

As hipóteses que tracei são as seguintes:

- O estado da amostra está relacionada ao tempo de utilização (horas)?

- O local/componente de aplicação influencia no resultado?

- Qual a influência do site de operação do equipamento na análise?

- Existe correlação entre o fabricante do lubrificante e o resultado da análise?

## Tipo de Problema

Este é um problema de **classificação supervisionada**.
Dado um conjunto de n características (C1, C2, C3...Cn), o objetivo é prever o resultado da análise da amostra do lubrificante, em função dos estados pré-definidos (Normal, Alerta, Crítico).

## Seleção de Dados

O dataset é extraído de sistema de análise laboratorial de óleos lubrificantes, alimentado pelos instrumentos de medição dos laboratórios. Alguns dados são inputados manualmente no sistema e os resultados são inseridos eletrônicamente, conforme as amostras são analisadas.

### Desvio Padrão

O desvio padrão é uma medida de dispersão que quantifica a quantidade de variação ou dispersão de um conjunto de valores. Um desvio padrão baixo indica que os pontos de dados tendem a estar próximos da média do conjunto, enquanto um desvio padrão alto indica que os pontos de dados estão espalhados por uma faixa maior de valores. Ele é a raiz quadrada da variância.

# Pré-Processamento de Dados

O pré-processamento de dados é uma etapa crucial para preparar os dados para modelagem, garantindo que estejam no formato correto e otimizados para o desempenho do algoritmo.

```python
# Separar features (X) e target (y)
X = df.drop('OVERALL_INTERP', axis=1)
y = df['OVERALL_INTERP']
```

```python
# Dividir os dados em conjuntos de treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
```

```python
print(f"Dimensões de X_train: {X_train.shape}")
print(f"Dimensões de X_test: {X_test.shape}")
print(f"Dimensões de y_train: {y_train.shape}")
print(f"Dimensões de y_test: {y_test.shape}")
```

## Normalização

A normalização escala os dados para um intervalo fixo, geralmente entre 0 e 1. É útil quando o algoritmo de machine learning assume que as características estão em uma escala semelhante.



```python
# Inicializar o MinMaxScaler
scaler_norm = MinMaxScaler()
```

```python
# Aprende min e max APENAS de X_train
scaler_norm.fit(X_train)
X_train_normalized = scaler_norm.transform(X_train)
# Usa a média e o desvio padrão aprendidos de X_train
X_test_normalized = scaler_norm.transform(X_test)
```

```python
# Exibir as primeiras linhas dos dados normalizados (como DataFrame para melhor visualização)
df_normalized = pd.DataFrame(X_train_normalized, columns=X_train.columns)
```

```python
print("\nPrimeiras 5 linhas dos dados normalizados (treino):")
print(df_normalized.head())
```

```python
# Visualização da distribuição após a normalização (exemplo para uma característica)
plt.figure(figsize=(8, 6))
sns.histplot(df_normalized['V100'], kde=True)
plt.title('Distribuição da Viscosidade (Normalizado)')
plt.xlabel('Viscosidade')
plt.ylabel('Frequência')
plt.show()
```

O histograma da *viscosidade (V100)* após a normalização mostra que os valores foram escalados para o intervalo de 0 a 1, mantendo a forma da distribuição original.

## Padronização

A padronização (ou Z-score scaling) transforma os dados para ter média 0 e desvio padrão 1. É útil para algoritmos que são sensíveis à escala das características, como SVMs ou redes neurais.

```python
# Inicializar o StandardScaler
scaler_std = StandardScaler()
```

```python
# Aprende média e desvio padrão APENAS de X_train
scaler_std.fit(X_train)
X_train_standardized = scaler_std.transform(X_train)
# Usa a média e o desvio padrão aprendidos de X_train
X_test_standardized = scaler_std.transform(X_test)
```

```python
# Exibir as primeiras linhas dos dados padronizados (como DataFrame para melhor visualização)
df_standardized = pd.DataFrame(X_train_standardized, columns=X_train.columns)
```

```python
print("\nPrimeiras 5 linhas dos dados padronizados (treino):")
print(df_standardized.head())
```

```python
# Visualização da distribuição após a padronização (exemplo para uma característica)
plt.figure(figsize=(8, 6))
sns.histplot(df_standardized['V100'], kde=True)
plt.title('Distribuição da Viscosidade (Padronizado)')
plt.xlabel('Viscosidade Padronizado')
plt.ylabel('Frequência')
plt.show()
```

O histograma da *viscosidade (V100)* após a padronização mostra que os valores foram transformados para ter uma média próxima de zero e um desvio padrão de um, centralizando a distribuição.
